```python
def fib(num):
    x,y=0,1
    for i in range(num):
        x,y=y,x+y
        yield x

def square(nums):
    for num in nums:
        yield num**2

x=square(5)  
print(next(x))
```

```python
# calculator
# CAlculator
terminate= False
while not terminate:
    n1= input("Enter the first number ")
    n2= input("Enter the second number ")
    n1 = int(n1)
    n2=int(n2)

    while True:
        operation = input('Please enter add/sub/mul/div or quiit to exit:')
        if operation== 'quit':
            terminate =True
            break
        if operation not in ['add','sub','mul','div']:
            print("Unknown operation")
            continue
        if operation == 'add':
            print('Sum of', n1 ,'and' ,n2 ,'is', n1+n2)
            break
        if operation == 'sub':
            print('sub of', n1 ,'and' ,n2 ,'is', n1-n2)
            break
        if operation == 'mul':
            print('mul of', n1 ,'and' ,n2 ,'is', n1 * n2)
            break
        if operation == 'div':
            print('div of', n1 ,'and' ,n2 ,'is', n1/n2)
            break
```

```python
"""
Data abstraction in python  or DATA HIDING IN PY
Abstraction is an important aspect of object-oriented programming.
In python, we can also perform data hiding by adding the double underscore (___) as a prefix to the attribute 
which is to be hidden. 
After this, the attribute will not be visible outside of the class through the object.
"""

class fruit:
    __count=0
    counts=0
    def __init__(self):
        fruit.__count=fruit.__count+1
    def display(self):
        print("The number of fruits : ",fruit.__count)
f1=fruit()

print(f1.counts)
try:
    print("hidden data",f1.__count)   ########## __count is hidden data
finally:
    f1.display()
    
```

```python
#multilevel inheritance
class food:
    def all(self):
        print("both veg and nonveg")

class vegfood(food):
    def veg(self):
        print("some people eat veg")

class nonvegfood(vegfood):
    def nonveg(self):
        print("some eat nonveg food")

nv=nonvegfood()
nv.nonveg()
nv.veg()
nv.all()
```

```python
#multiple inheritance

#Python provides us the flexibility to inherit multiple base classes in the child class.
'''
class Base1:  
    <class-suite>  
class Base2:  
    <class-suite>  
.  
.  
class BaseN:  
    <class-suite>  
class Derived(Base1, Base2, ...... BaseN):  
    <class-suite>  
'''
class calculation1:
    def add(self,a,b):
        return a+b
class calculation2:
    def sub(self,a,b):
        return a-b
class calculation3:
    def mul(self,a,b):
        return a*b
class baap(calculation1,calculation2,calculation3):
    def divide(self,a,b):
        return a//b

b1=baap()
print(b1.divide(8,4))
print(b1.add(8,4))
print(b1.sub(8,4))
print(b1.mul(8,4))
```
